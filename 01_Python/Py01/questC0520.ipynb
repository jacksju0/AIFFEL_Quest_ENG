```python
####################
# 풀이
####################

def is_palindrome(string):
    reversed_str=string[::-1]
    if string == reversed_str:
        return True
    else:
        return False
```

### Q1-2 회문 분류기

```python
# 우선 Q1-2 의 문제 1번 부터 풀자
# 일단 함수 안에서 딕셔너리를 만들어야 겠네?
# 딕셔너리의 키는 "회문O" 와 "회문X" 두 개,
# 그리고 함수에서 입력으로 받은 리스트를 하나씩 까면서
# 회문은 "회문O"에 넣고, 아니면 회문X에 넣으면 되겠네?

def palindrome_classifier(str_list):
    dict_classifier={'회문o':[],'회문x':[]}
    
    for str1 in str_list:
        if is_palindrome(str1)==True:
            dict_classifier['회문o'].append(str1)
        else:
            dict_classifier['회문x'].append(str1)
    return dict_classifier
```

```python
####################
# 테스트
####################
targets=['기러기','까마귀','토마토','스윗소스','스위스','인도인','한국인','별똥별','우영우','우향우']
print(palindrome_classifier(targets))
```

```python
# 이제 Q1-2의 문제 2를 추가 구현하자
# 리스트의 총 요소수 = 리스트의 길이 = len(리스트 변수 이름)
# 어? 문제를 자세히 보니 함수의 최종 출력 형태가 () ? 튜플이네?
# 괄호를 안해줘도 상관없지만 가독성을 위해 넣자.(오해나 착오를 줄인다)
####################
# 풀이
####################
def palindrome_classifier(str_list):
    dict_classifier={'회문o':[],'회문x':[]}
    
    for str1 in str_list:
        if is_palindrome(str1)==True:
            dict_classifier['회문o'].append(str1)
        else:
            dict_classifier['회문x'].append(str1)

    num_total=len(str_list)
    num_palindrome=len(dict_classifier['회문o'])
    
    return (dict_classifier, num_palindrome, num_palindrome/num_total)
```

```python
#잘 되긴하는데... 왠지 이 코드의 for문을 한 줄로 만들 수 있을것 같은데...
#한 줄로 만들긴 했는데 가독성은 좀 떨어지는것 같네.
######################
# 개선 코드
######################

def palindrome_classifier(str_list):
    dict_classifier = {'회문o': [], '회문x': []}
    
    for str1 in str_list:
        # 조건에 따라 append할 target 리스트를 결정한 후 한 줄로 처리
        dict_classifier['회문o' if is_palindrome(str1) else '회문x'].append(str1)
            
    num_total=len(str_list)
    num_palindrome=len(dict_classifier['회문o'])
    
    return (dict_classifier, num_palindrome, num_palindrome/num_total)
```

### Q2-1 성적 데이터화

```python
# 문제 Q2-2를 보니까
# Q2-1을 작은 데이터베이스로 만들어서
# 이름을 호출하면 그 사람과 관련된 정보가 나와야 하네?
# 음... 딕셔너리를 써도 되는데
#
# 여기서 중요한건 정보를 불러와서 이것 저것 계산을 한다는거네
# 그리고 계산은 표를 보면 표 전체 데이터, 행 방향 데이터, 열방향 데이터
# 이렇게 규칙성 있게 쓰네?
# 그럼 숫자 데이터만 2차원 배열로 만들어야 겠다.
# 그리고 각 요소에 접근 할 수 있는 번호를 변수로 만들어 두자

####################
# 풀이
####################
subjects=("국어","영어","수학")
name=("유원영","설지민","신윤아","장윤아")
scores=[[94,91,89],
        [88,96,90],
        [86,88,77],
        [98,76,92]]
```

### Q2-2 성적 분석기

```python
# 이름으로 딕셔너리를 불러오고,
# 각 과목정보는 튜플의 위치를 가지고 list에 접근하면 되고
# 리스트의 숫자합은 sum(리스트 변수)를 하면 한방에 구할 수 있지..
# 그런데 문제는 지금 2차원 배열은 이름으로 묶여 있다는거지
# sum(유원영)은  가능하지만, sum(국어)는 안되는 상황이지.
# 내가 필요한건 과목 단위로 묶여있는건데.
# 2차원 배열을 뒤집어서 다시 묶을 수는 없나?
# 찾아보니까 있네?
scores_by_subject = list(zip(*scores))
print(scores_by_subject)
```

```python
#그럼 각 과목의 학급 평균은
idx=subjects.index("국어")
sum(scores_by_subject[idx])/len(name)

#이렇게 하나씩 하는 것 보다 for문을 돌리자
avg_class_subject=[]
for idx in range(len(subjects)):
    # idx가 0일 때는 국어, 1일 때는 영어, 2일 때는 수학 점수
    # 결과도 배열로 저장
    current_avg = sum(scores_by_subject[idx]) / len(name)
    avg_class_subject.append(current_avg)

    subject_name = subjects[idx]
    print(f"{subject_name} 학급 평균: {avg_class_subject[idx]:.2f}")

# 전과목 학습 평균은
# avg_class_total=(3과목 평균 합)/len(subjects)
avg_class_total=sum(avg_class_subject)/len(subjects)
print(f"전과목 학급 평균: {avg_class_total:.2f}")
```

```python
#모든 준비는 끝났다 함수로 만들어 보자
def score_analysis(student):
    #============================================================
    # 위에서 구한 과목별, 전과목 학습 평균
    #============================================================
    avg_class_subject=[]
    for idx in range(len(subjects)):
        # idx가 0일 때는 국어, 1일 때는 영어, 2일 때는 수학 점수
        # 결과도 배열로 저장
        current_avg = sum(scores_by_subject[idx]) / len(name)
        avg_class_subject.append(current_avg)

    # 전과목 학습 평균은
    # avg_class_total=(3과목 평균 합)/len(subjects)
    avg_class_total=sum(avg_class_subject)/len(subjects)
    #============================================================

    #============================================================
    # name 에 대한 정보 처리
    #============================================================
    print(student)
    #일단 국어 한과목에 대한것만 뜨게 해보자
    idx=0
    subject_name=subjects[idx]                 # 과목명 ("국어", "영어", "수학")
    student_idx=name.index(student)
    student_score=scores[student_idx][idx]     # 해당 학생의 과목 점수
    class_avg=avg_class_subject[idx]           # 이전에 구해둔 과목별 학급 평균
    score_diff = student_score - class_avg     # 학급 평균 대비 차이 계산 (내 점수 - 학급 평균)
    print(f"{subject_name}: {student_score}점, 학급 평균({class_avg:.2f}점) 대비 {score_diff}")

    #그리고 전과목에 대해서도 출력해보자
    avr_student_total=sum(scores[student_idx])/len(subjects)
    score_diff = avr_student_total - avg_class_total     # 학급 평균 대비 차이 계산 (내 점수 - 학급 평균)
    print("")
    print(f"전과목: 평균: {avr_student_total:.2f}점, 학급 평균({avg_class_total:.2f}점) 대비 {score_diff:.2f}")
```
